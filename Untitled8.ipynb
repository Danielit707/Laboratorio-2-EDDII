```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from math import radians, sin, cos, sqrt, atan2

# Cargar el archivo CSV
path = "/content/drive/MyDrive/dataset/flights_final.csv"
df = pd.read_csv(path)

# Función para calcular la distancia Haversine
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Radio de la Tierra en km
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat / 2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Crear el grafo
G = nx.Graph()

def obtener_posiciones_por_coordenadas():
    posiciones = {}
    for index, row in df.iterrows():
        codigo_origen = row['Source Airport Code']
        lat1, lon1 = row['Source Airport Latitude'], row['Source Airport Longitude']
        if codigo_origen not in posiciones:
            posiciones[codigo_origen] = (lon1, lat1)

        codigo_destino = row['Destination Airport Code']
        lat2, lon2 = row['Destination Airport Latitude'], row['Destination Airport Longitude']
        if codigo_destino not in posiciones:
            posiciones[codigo_destino] = (lon2, lat2)
    return posiciones

def obtener_posiciones_escaladas(separacion=2.5):
    posiciones = {}
    for index, row in df.iterrows():
        codigo_origen = row['Source Airport Code']
        lat1, lon1 = row['Source Airport Latitude'], row['Source Airport Longitude']
        if codigo_origen not in posiciones:
            posiciones[codigo_origen] = (lon1 * separacion, lat1 * separacion)

        codigo_destino = row['Destination Airport Code']
        lat2, lon2 = row['Destination Airport Latitude'], row['Destination Airport Longitude']
        if codigo_destino not in posiciones:
            posiciones[codigo_destino] = (lon2 * separacion, lat2 * separacion)
    return posiciones

# Añadir nodos y aristas al grafo con las distancias calculadas entre los aeropuertos
for index, row in df.iterrows():
    lat1, lon1 = row['Source Airport Latitude'], row['Source Airport Longitude']
    lat2, lon2 = row['Destination Airport Latitude'], row['Destination Airport Longitude']
    distance = haversine(lat1, lon1, lat2, lon2)

    G.add_edge(row['Source Airport Code'], row['Destination Airport Code'], weight=distance)

# Función para mostrar el menú de opciones
def mostrar_menu():
    print("\n----- Menú de Opciones -----")
    print("1. Verificar si el grafo es conexo")
    print("2. Determinar el peso del Árbol de Expansión Mínima (MST)")
    print("3. Mostrar la información de un aeropuerto")
    print("4. Mostrar el camino mínimo entre dos aeropuertos")
    print("5. Visualizar el grafo de rutas aéreas")
    print("6. Salir")
    return input("Elige una opción (1-6): ")

# 1. Verificar si el grafo es conexo
def verificar_conexo():
    is_conexo = nx.is_connected(G)
    if not is_conexo:
        componentes = list(nx.connected_components(G))
        print(f"El grafo no es conexo. Hay {len(componentes)} componentes.")
        for idx, comp in enumerate(componentes, 1):
            print(f"Componente {idx}: {len(comp)} vértices.")
    else:
        print("El grafo es conexo.")
        print(f"Vértices en el grafo: {G.number_of_nodes()}")

# 2. Determinar el peso del Árbol de Expansión Mínima (MST)
def peso_mst():
    if nx.is_connected(G):
        mst = nx.minimum_spanning_tree(G)
        peso_mst = mst.size(weight='weight')
        print(f"El peso del Árbol de Expansión Mínima es: {peso_mst:.2f} km.")
    else:
        componentes = list(nx.connected_components(G))
        for idx, comp in enumerate(componentes, 1):
            subgrafo = G.subgraph(comp)
            mst = nx.minimum_spanning_tree(subgrafo)
            peso_mst = mst.size(weight='weight')
            print(f"Peso del Árbol de Expansión Mínima del Componente {idx}: {peso_mst:.2f} km.")

# 3. Mostrar la información de un aeropuerto y los 10 aeropuertos más lejanos
def mostrar_info_aeropuerto(codigo, lejanos):
    # Buscar tanto en Source como en Destination
    info_source = df[df['Source Airport Code'] == codigo]
    info_dest = df[df['Destination Airport Code'] == codigo]

    if not info_source.empty:
        info = info_source.iloc[0]
    elif not info_dest.empty:
        info = info_dest.iloc[0]
    else:
        print(f"No se encontró información del aeropuerto con código {codigo}.")
        return

    print(f"\nInformación del Aeropuerto ({codigo}):")
    print(f"Nombre: {info['Source Airport Name']}")
    print(f"Ciudad: {info['Source Airport City']}")
    print(f"País: {info['Source Airport Country']}")
    print(f"Latitud: {info['Source Airport Latitude']}")
    print(f"Longitud: {info['Source Airport Longitude']}")
    if (lejanos):
      # Calcular los caminos más largos desde este aeropuerto
      caminos_minimos = nx.single_source_dijkstra_path_length(G, codigo, weight='weight')
      caminos_largos = sorted(caminos_minimos.items(), key=lambda x: x[1], reverse=True)[:10]

      print("\n10 aeropuertos más lejanos:")
      for i, (aeropuerto, distancia) in enumerate(caminos_largos, 1):
          info_aeropuerto = df[(df['Source Airport Code'] == aeropuerto) | (df['Destination Airport Code'] == aeropuerto)].iloc[0]
          print(f"{i}. Código: {aeropuerto}, Distancia: {distancia:.2f} km, Ciudad: {info_aeropuerto['Source Airport City']}, País: {info_aeropuerto['Source Airport Country']}")


# 4. Encontrar y mostrar el camino mínimo entre dos aeropuertos, y mostrarlo en el mapa
def camino_minimo(codigo_origen, codigo_destino):
    try:
        camino = nx.shortest_path(G, source=codigo_origen, target=codigo_destino, weight='weight')
        distancia_total = nx.shortest_path_length(G, source=codigo_origen, target=codigo_destino, weight='weight')

        print(f"\nEl camino mínimo entre {codigo_origen} y {codigo_destino} es:")
        for aeropuerto in camino:
            mostrar_info_aeropuerto(aeropuerto, False)
        print(f"\nDistancia total: {distancia_total:.2f} km.")

        # Visualizar el camino en el mapa
        visualizar_camino(camino)
    except nx.NetworkXNoPath:
        print(f"No hay camino entre {codigo_origen} y {codigo_destino}.")

# 5. Visualizar el grafo completo
def visualizar_grafo():
    pos = obtener_posiciones_por_coordenadas()  # Usar posiciones geográficas
    plt.figure(figsize=(24, 24))

    # Dibujar nodos y aristas
    nx.draw(G, pos, with_labels=True, node_size=300, node_color="skyblue", font_size=10, font_weight="bold", edge_color="gray")

    # Añadir etiquetas de pesos (distancias) a las aristas
    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)

    # Mostrar el grafo
    plt.title("Grafo de rutas aéreas")
    plt.show()


# Visualizar un camino específico en el mapa
def visualizar_camino(camino):
    pos = nx.spring_layout(G)  # Generar una disposición para los nodos
    plt.figure(figsize=(24, 24))

    # Dibujar nodos y aristas
    nx.draw(G, pos, with_labels=True, node_size=300, node_color="lightblue", font_size=10, font_weight="bold", edge_color="gray")

    # Resaltar el camino en el grafo
    camino_edges = list(zip(camino[:-1], camino[1:]))
    nx.draw_networkx_edges(G, pos, edgelist=camino_edges, edge_color='r', width=2)

    # Añadir etiquetas de pesos (distancias) a las aristas del camino
    edge_labels = {(u, v): f"{G[u][v]['weight']:.2f} km" for u, v in camino_edges}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red', font_size=8)

    # Mostrar el grafo
    plt.title(f"Camino mínimo entre {camino[0]} y {camino[-1]}")
    plt.show()

# Función principal del programa
def main():
    while True:
        opcion = mostrar_menu()

        if opcion == '1':
            verificar_conexo()
        elif opcion == '2':
            peso_mst()
        elif opcion == '3':
            codigo = input("Ingresa el código del aeropuerto: ").upper()
            if codigo in G:
                mostrar_info_aeropuerto(codigo, True)
            else:
                print("Código de aeropuerto no válido.")
        elif opcion == '4':
            codigo_origen = input("Ingresa el código del aeropuerto de origen: ").upper()
            codigo_destino = input("Ingresa el código del aeropuerto de destino: ").upper()
            if codigo_origen in G and codigo_destino in G:
                camino_minimo(codigo_origen, codigo_destino)
            else:
                print("Uno o ambos códigos de aeropuertos no son válidos.")
        elif opcion == '5':
            visualizar_grafo()
        elif opcion == '6':
            print("Saliendo del programa.")
            break
        else:
            print("Opción no válida. Intenta de nuevo.")

# Ejecutar el programa
main()
```
